```python
# Incriment ID values (to avoid keeping them as zeros)
def fillNA(data):
    """Function to Incriment ID values"""
    data = np.array(data)
    # Convert to float
    toFloat = [float(floatValue) for floatValue in data]       
    # Find biggest id available (to avoid overlapping)
    biggestId = np.max(toFloat)                                
    print("Max id number available =", biggestId)
    # Increment ID value
    for i in range (len(data)):
        if float(data[i]) == 0:
            data[i] = biggestId   
            biggestId = biggestId +1                           
    return data
```

### ML Metrics

```python
print(cancellation)
print(no_cancellation)
```

Our target variable is cancellation. A cancellation occurs 32.6% of the time and maintained bookings 67.4% of the time.

Here is how our metrics can be interpreted.

True Positive: Correctly predicting a cancellation.

False Positive: Incorrectly predicting a cancellation when customer maintains booking.

True Negative: Correctly predicting a maintained booking when customer does not cancel.

False Negative: Incorrectly predicting a maintained booking when customer does cancel.

Accuracy (TP+TN/TP+FP+FN+TN): Accuracy is the ratio of correctly predicted observations to the total observations. This is the ratio of correctly cancellations, out of the total population of bookings.

Precision (TP/TP+FP): Precision is the ratio of correctly predicted positive observations to the total predicted positive observations. This is the ratio of correctly predicted cancellations out of all predicted cancellations (some of which were incorrect predictions and in actuality maintained bookings).

Recall (TP/TP+FN): Recall is the ratio of correctly predicted positive observations to the all observations in actual class. Or, all correctly classified cancellations, out of all actual cancellations (some of which were missed and marked as maintained bookings).

F1 (2x(Recall x Precision) / (Recall + Precision)): F1 Score is the weighted average of Precision and Recall, thereby taking both false positives and false negatives into account. F1 ranges from 0 to 1, with a better score being higher.

Which metric matters? 

In this project, I am interested in predicting cancellations early to allow time for hotel management to find alternate buyers for empty rooms. To allow time to proactively find other customers, I would want to minimize false negatives or unanticipated cancelled bookings which could be avoided. To measure my models, I will use the F1 score since it accounts for both precision and recall. I want my F1 score to be as high as possible out of 1.

## K-means Clusters

K-means is a method of vector quantization, which partitions n observations into k clusters in which each observation belongs to the cluster with the nearest mean.

```python
X.head(3)
```

Below I am deliberately using standard scaling instead of minmax scaling because the standard scaler does a z-score transformation and allows the data to have a bigger spread and therefore a bigger difference between the clusters. When I tried minmax scaling, the difference was not adequate to perform clustering. 

```python
# Scale values to make sure they are weighed equally
scaler = StandardScaler()
scaler.fit(X)
scaled_X = scaler.transform(X)
X1 = pd.DataFrame(scaled_X)
X1.columns = X.columns
X1.describe()
```

Intuitively, inertia tells you how far away from a cluster the surrounding points are. Therefore, we want inertia to be as low as possible so the clusters are concentrated. In the below chart, there is no distinct "elbow", but inertia appears to be optimal between 4 - 10 clusters.

```python
# Fit multiple k-means algorithms and store values in an empty list
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster, init='k-means++')
    kmeans.fit(X1)
    SSE.append(kmeans.inertia_)

# Convert results into a dataframe and plot them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
# Clustering (without PCA, for interpretability)

# Configure clustering algorithm to run 10x and generate clusters
kmeans = KMeans(init='k-means++', n_clusters=4, n_init=10)
# Apply clustering
kmeans.fit(X1)
# Inertia on the fitted data
print(kmeans.inertia_)
# Predict clusters
y_kmeans = kmeans.predict(X1)
print(y_kmeans)
```

```python
# Function to train input classifier using training data, and return time to train
def train_classifier(clf, X_train, y_train):
    print("Training using {}".format(clf.__class__.__name__))
    clf.fit(X_train, y_train)
    start = time.time()
    end = time.time()
    print("Completed training classifier with training data in {:.3f} secs".format(end - start))
    return (end - start)

# Set desired classifier
clf=DummyClassifier()

# Apply training method
print('\nTraining:')
train_classifier(clf, X_train_cv, y_train_cv) 

# Function to use input classifier to predict training set and compute F1 score
def predict_labels(clf, features, target):
    y_pred = clf.predict(features)
    start = time.time()
    end = time.time()
    print ("Completed predicting class in {:.3f} secs".format(end - start))
    # Take F1 between actual target values and predicted from classifier
    return f1_score(target.values, y_pred, average='micro') , (end - start)
```

### Compare Classifiers

The goal below is to train the 5 classifiers on the training dataset and then predict which clusters new data belongs to. 

To measure their performance, I will use training time, prediction time of training data, f1 score of training data, prediction time of test data and f1 score of test data. 

From there, I will select the best 3 classifiers, and optimize their performance by using grid search for the best hyperparemeters.

```python
# To iterate classification over 5 models
noOfTypes = 5;

# Create table containing specified columns with corresponding string and float32 data types
t1 = Table(names=('Type','Training time' ,
                  'Prediction time-Training', 
                  'F1 score-Training', 
                  'Prediction time-Testing', 
                  'F1 score-Testing'), 
           dtype=('S25', 'f4','f4', 'f4','f4', 'f4'))

"""Table containing performance scores of different classifiers"""

for i in range(0,noOfTypes):
    if i==0:    
        clf = DummyClassifier()
    if i==1:    
        clf = DecisionTreeClassifier(random_state=0)
    if i==2:    
        clf = RandomForestClassifier()
    if i==3:    
        clf = LogisticRegression()
    if i==4:    
        clf = KNeighborsClassifier()
        
    # Class  
    clfType=clf.__class__.__name__;
    # Return time to train classifier
    trainTime=train_classifier(clf, X_train_cv, y_train_cv);
    
    # Vector to return output 0 and 1 in prediction function from training data
    retVec=predict_labels(clf, X_train_cv, y_train_cv)
    trainF1Score=retVec[0];
    trainTime=retVec[1];

    # Vector to return output 0 and 1 in prediction function from testing data
    retVec1=predict_labels(clf, X_test_cv, y_test_cv)
    testF1Score=retVec1[0];
    testTime=retVec1[1];
    
    # Add column names to table
    t1.add_row((clfType,trainTime,trainTime, trainF1Score,testTime, testF1Score))

# Show table
t1    
```
